## Cleaning data

### 1. Coffees should contain only numeric values
#### Note: Surprisingly Nan are valid numbers

```python
# we should convert string values to null if any 
# we can mention errors = 'coerce' to do that
data.coffees = pd.to_numeric(data.coffees, errors = 'coerce')
data.head()
```

let's see the type of 'coffees' column now

```python
# coffees column is of Series type .
# Values for this attribute are of float64 type
print(type(data.coffees[0]))   
print(type(data.coffees))
```

#### got rid of textual values but what to do with NaNs
#### We can 1. drop the corresponding rows  or 2. fill up NaNs with mean of adjacent values

```python
# Use .dropna()
data.dropna()
data.head()
```

nothing happend !!!!! 
Well Pandas works on copies of objects for safety bcoz you don't want to perform things inplace  
If you do want so mention inplace = True


```python
data.dropna(inplace = True)
# this is similar to
# data = data.dropna()  .... I hope you are getting it
data.head()
```

You can see that rows with index 2 and 3 are droppes now.  
Note: Pandas does maintain index for objects

Now it is of float64 type but we want it to be of int type

```python
data.coffees = data.coffees.astype(int)  # type changed and assigned to original
print(type(data.coffees))
data.head()
```

### 2. Timestamp should be of 'datetime' type

```python
data.timestamp = pd.to_datetime(data.timestamp)
print(type(data.timestamp[0]))
data.dtypes
```

#### 1. We can make our index as our timestamp and dropo timestamp col

```python
data.index
```

```python
data.index = data.timestamp
data.index
```

```python
data.head()
```

#### See our index got changed to timestamp but we still have timestamp column

```python
# .drop() takes list ob objects(rows) to drop from dataframe
data.drop(['timestamp'],axis=1,inplace = True)
# axis = 1 for columnwise
# axis = 0 for rowwise (deault)
data.head()
```

#### Let's add some rows at midnight on each day

```python
# pd.date_range() with daily frew and normalization
midnights = pd.date_range(data.index[0],data.index[-1],
                         freq = 'D',normalize = True)
# normalize means all timestamp value normalized to midnight 
# freq is a Day ie. per day
# So we are taking the timestamp(midnight ones) from the range of our index per day
midnights
```

take union with our index so that we will have one entry at midnight for each day

```python
new_index = midnights.union(data.index)
new_index
```

ok now we've got one entry at midnight for each day in the range of our timestamp

#### now reindex our data with this new_index

```python
unsampled_data = data.reindex(new_index)
# notice we have added only the timestamp(midnight timestamp entry) but we haven't provide other values 
unsampled_data.head()
```

#### See we have midnight entry for each day

#### We can fill those NaNs using interpolation 
But remember our data is not uniformly spaced acc to time   .. so we can plot data against time and take the value which corresponds to the midnight

```python
unsampled_data = unsampled_data.interpolate(method = 'time')
unsampled_data.head(10)
# when you apply interpolate() on dataframe using method = 'time'  it looks for time objects in index 
# if index is of time objects, then plot each column against those time objects and fill the empty values
# ie. the x-axis is time value 
```

#### Let's resample the data at a daily freq

```python
daily_data = unsampled_data.resample('D').count()
```

```python
daily_data
```

it is giving count of coffees for each day now. 
Also we dont't want contributor col

```python
daily_data = daily_data.drop(['contributor'],axis = 1)
daily_data
```

Let's resample daily but only take midnight values


```python
daily_data = unsampled_data.resample('D').asfreq()
# asfreq() - we are telling that reasample on daily basis and take the midnight value
daily_data = daily_data.drop(['contributor'],axis =1)
daily_data
```

We can get weekday names 

```python
# generate a column of weekday_names
daily_data['weekday'] = daily_data.index.weekday_name
daily_data.head()
```

#### Previously for geting a weekday name we wrote a function but guess what pandas can do that for us. I hope u saw weekday_name attribute to the index(which is of datetime objects)

Now we have uniformly sampled data across time (ie. midnight for each day)

## How many coffees were made on  a certain day ?

```python
# use .diff() 
# diff() gives the difference of two adjacent values
coffees_made = daily_data.coffees.diff()
coffees_made.head()
```

Let's shift this data up 

```python
coffees_made = daily_data.coffees.diff().shift(-1)
coffees_made.head(10)
```

so on 4th 34 coffees were made, on 5th 55 coffees were made

### Let's add this to our data as another column - coffees_today

```python
daily_data["coffeees_today"] = coffees_made
daily_data.head(10)
```

 ### Average coffees_made on paricular day of week
 Let's groupby weekday

```python
coffees_by_day = daily_data.groupby("weekday").mean()
coffees_by_day
```
